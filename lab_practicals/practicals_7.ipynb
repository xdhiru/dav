<H2>Q7. Consider the following data frame containing a family name, gender of the family member and her/his monthly income in each record. Write a program in Python using Pandas to perform the following:</H2>

```python
import pandas as pd
import numpy as np
src="/content/drive/MyDrive/Classroom/DAV sem 3/16027_DHIRENDRA_KUMAR_PATEL_DAV/lab_practicals/csvdata"
```

```python
fam1=pd.read_csv(f"{src}/q7.csv")
fam1
```

<H3>a. Calculate and display familywise gross monthly income.</H3>

```python
print("Familywise gross monthly income : ")
fam1.groupby('FamilyName')['MonthlyIncome(Rs.)'].sum()
```

<H3>b. Display the highest and lowest monthly income for each family name</H3>

```python
print("Highest income for each family :")
fam1.groupby('FamilyName')['MonthlyIncome(Rs.)'].max()
```

```python
print("Highest income for each family :")
fam1.groupby('FamilyName')['MonthlyIncome(Rs.)'].min()
```

<H3>c. Calculate and display monthly income of all members earning income less than Rs. 80000.00</H3>

```python
print("Monthly income of all members earning less than Rs. 80000.00 :")
fam1[fam1["MonthlyIncome(Rs.)"]<80000]
```

<H3>d. Display total number of females along with their average monthly income</H3>

```python
fam1[fam1["Gender"]=="Female"].groupby("FamilyName").agg([('Female count', 'count'),('Average Income', 'mean')])
```


<h3>e. Delete rows with Monthly income less than the average income of all members</h3>

```python
fam1[fam1["MonthlyIncome(Rs.)"]>fam1["MonthlyIncome(Rs.)"].mean()]
```
