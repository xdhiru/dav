###c. Plot bar chart to show the frequency of each class label in the data.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
target_names = iris.target_names
iris_target = iris.target

# Calculate the frequency of each class label
label_counts = [len(iris_target[iris_target == i]) for i in range(len(target_names))]

# Create a bar chart using Matplotlib
plt.bar(target_names, label_counts, color=['skyblue', 'lightgreen', 'lightcoral'])

# Add labels and title
plt.xlabel('Class Label')
plt.ylabel('Frequency')
plt.title('Frequency of Each Class Label in Iris Dataset')

# Show the plot
plt.show()
```

###f. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Create a pair plot using Matplotlib and Pandas
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
scatter_matrix = pd.plotting.scatter_matrix(
    iris_df, figsize=(10, 10), diagonal='hist', c=iris_df['species'].apply(lambda x: colors[x]), marker='o', alpha=0.8
)

# Add legends
for ax in scatter_matrix.ravel():
    ax.set_xlabel(ax.get_xlabel(), fontsize=10)
    ax.set_ylabel(ax.get_ylabel(), fontsize=10)
    ax.tick_params(axis='both', which='both', length=0)
    ax.legend(*ax.get_legend_handles_labels(), fontsize=8)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

###g. Draw heatmap for any two numeric attributes


```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate the correlation matrix
correlation_matrix = np.corrcoef(iris_df.T)

# Create a heatmap using Matplotlib
plt.figure(figsize=(10, 8))
heatmap = plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(heatmap, fraction=0.046, pad=0.04)

# Display feature names as labels on both axes
plt.xticks(np.arange(4), iris.feature_names, rotation=45)
plt.yticks(np.arange(4), iris.feature_names)

# Show the plot
plt.show()
```

###i. Compute correlation coefficients between each pair of features and plot heatmap

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_data = iris.data

# Calculate correlation matrix
correlation_matrix = np.corrcoef(iris_data.T)

# Create a heatmap using Matplotlib
plt.figure(figsize=(10, 8))
heatmap = plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')

# Add colorbar
plt.colorbar(heatmap, fraction=0.046, pad=0.04)

# Display feature names as labels on both axes
feature_names = iris.feature_names
plt.xticks(np.arange(len(feature_names)), feature_names, rotation=45)
plt.yticks(np.arange(len(feature_names)), feature_names)

# Add title
plt.title('Correlation Coefficients Between Iris Features')

# Show the plot
plt.show()
```
