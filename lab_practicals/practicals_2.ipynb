<H3>b. Create a series with N elements with some duplicate values. Find the minimum and maximum ranks assigned to the values using ‘first’ and ‘max’ methods</H3>

```python
s2=pd.Series([10,20,30,60,20,30])
s2
```

```python
s2.rank(method='first')
```

```python
s2.rank(method='max')
```

<H3>c. Display the index value of the minimum and maximum element of a Series</H3>

```python
print("minimum : ", s2.min())
print("index of minimum : ",s2.idxmin())
```

```python
print("maximum : ", s2.max())
print("index of maximum : ",s2.idxmax())
```
