<H3>c. Test whether the elements of a given 1D array are zero, non-zero and NaN. Record the indices of these
elements in three separate arrays. </H3>

```python
arr3=np.array([np.NaN,0,3,0,np.NaN,4])
print("Given 1-D Array :",arr3)
zero,nonzero,nan=[],[],[]
for x in range(len(arr3)):
    if arr3[x]==0:
      zero.append(x)
    if np.isnan(arr3[x]):
      nan.append(x)
    else:
      nonzero.append(x)
print("\nZero Indexes :",np.array(zero))
print("\nNaN Indexes :",np.array(nan))
print("\nNon-Zero Indexes :",np.array(nonzero))
```

<H3>d. Create three random arrays of the same size: Array1, Array2 and Array3 Subtract Array 2 from Array3 and store in Array4. Create another array Array5 having two times the values in Array1. Find Covariance and Correlation of Array1 with Array4 and Array5 respectively.</H3>

```python
array1=np.random.randint(10, size=(3, 3))
array2=np.random.randint(10, size=(3, 3))
array3=np.random.randint(10, size=(3, 3))
array4=array3-array2
array5=array1*2
print("Covarience of Array1 with Array 4 :\n", np.cov(array1, array4))
print("\nCovarience of Array1 with Array 5 :\n", np.cov(array1, array5))
print("\nCorrelation of Array1 with Array 4 :\n", np.corrcoef(array1, array4))
print("\nCorrelation of Array1 with Array 5 :\n", np.corrcoef(array1, array5))
```

<h3>e. Create two random arrays of the same size 10: Array1, and Array2. Find the sum of the first half of both the arrays and product of the second half of both the arrays.</h3>

```python
arr1=np.random.randint(100,size=10)
arr2=np.random.randint(100,size=10)
print("Array 1 :\n", arr1)
print("\nArray 2 :\n", arr2)

print("\nSum of first half of both the arrays :\n", np.sum(arr1[:5])+np.sum(arr2[:5]))
print("\nProduct of second half both the arrays :\n", np.prod(arr1[5:])*np.prod(arr2[5:]))
```

<h3>g. Create a 2-dimensional array of size m x n having integer elements in the range (10,100). Write statements to swap any two rows, reverse a specified column and store updated array in another variable.</h3>

```python
m=int(input("Enter vaue of m :"))
n=int(input("Enter vaue of n :"))
arr8=np.random.randint(10,100, size=(m,n))
print("\nArray :\n",arr8)

key=1
while(key!=0):
  key=int(input("\n===============\n0. Exit \n1. Swap any two rows.\n2. Reverse a specified column.\n\nEnter choice (0-2) :"))
  if key==1:
    srow1=int(input(f"\nEnter row to be swapped [0-{m-1}] : "))
    srow2=int(input(f"\nEnter row to swap with [0-{m-1}] : "))
    temp=np.array(arr8[srow1])
    arr8[srow1]=arr8[srow2]
    arr8[srow2]=temp
    print("\nArray after swapping rows :\n",arr8)

  elif key==2:
    rcol=int(input(f"\nEnter column to be reversed [0-{n-1}]:"))
    arr9=np.array(arr8)
    arr9[:,rcol]=arr9[::-1,rcol]
    print("\nArray after reversing column :\n",arr9)

  elif key==0:
    break
```
