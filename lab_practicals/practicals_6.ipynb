###a. Clean the data by dropping the column which has the largest number of missing values.

```python
column_most_missing = ship.isnull().sum().idxmax()
ship.drop(column_most_missing, axis=1)
```

###b. Find total number of passengers with age more than 30

###c. Find total fare paid by passengers of second class

```python
ship[ship['Pclass'] == 2]['Fare'].sum()
```

###d. Compare number of survivors of each passenger class

```python
ship.groupby("Pclass")["Survived"].agg([("Survivors","sum")])
```

###g. Compare density distribution for features age and passenger fare

```python
import pandas as pd
import matplotlib.pyplot as plt

# Density distribution for 'age'
plt.subplot(1, 2, 1)
plt.hist(ship['Age'].dropna(), bins=30, density=True, color='skyblue', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Density Distribution of Age in Titanic')

# Density distribution for 'fare'
plt.subplot(1, 2, 2)
plt.hist(ship['Fare'], bins=30, density=True, color='lightcoral', alpha=0.7)
plt.xlabel('Fare')
plt.ylabel('Density')
plt.title('Density Distribution of Fare in Titanic')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

###h. Draw the pie chart for three groups labelled as class 1, class 2, class 3 respectively displayed in different  colours. The occurrence of each group converted into percentage should be displayed in the pie chart.  Appropriately Label the chart.

```python
import matplotlib.pyplot as plt

Pclasses=ship.groupby("Pclass").count()
class1=Pclasses.iloc[0,:][0]
class2=Pclasses.iloc[1,:][0]
class3=Pclasses.iloc[2,:][0]

group_counts = [class1,class2,class3]
# Labels for each group
labels = ['Class 1', 'Class 2', 'Class 3']

# Colors for each group
colors = ['skyblue', 'lightgreen', 'lightcoral']

# Calculate percentages
total = sum(group_counts)
percentages = [(count / total) * 100 for count in group_counts]

# Create a pie chart using Matplotlib
plt.pie(percentages, labels=labels, colors=colors)

# Add title
plt.title('Distribution of Three Classes')
plt.legend()
plt.tight_layout()
plt.show()
```
