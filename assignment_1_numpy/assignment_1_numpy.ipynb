##2. Explain Broadcasting in numpy and Create 2 dimension array with user input and validate for implementation of broadcasting on arithmetic operation (+,-,*,/,%). Show proper results on every operation.

###Ans: Broadcasting in NumPy allows to perform element-wise operations on arrays of different shapes and sizes and eliminates the need for explicit looping over array elements.

```python
import numpy as np
rows=int(input("Enter no. of rows :"))
cols=int(input("Enter no. of columns :"))
arr2=np.zeros((rows,cols),dtype=float)
for x in range(rows):
  for y in range(cols):
    arr2[x][y]=float(input(f"Enter the element at {x},{y} :"))
print("Array : \n",arr2)
print("\nBroadcast Addition (+5) : \n",arr2+5)
print("\nBroadcast Subtraction (-3) : \n",arr2-3)
print("\nBroadcast Multiplication (*10) : \n",arr2*10)
print("\nBroadcast Division (/5) : \n",arr2/5)
print("\nBroadcast Modulo(returns remainder) (%3) : \n",arr2%3)
```

##6. Create a 2 dimensional array using user input and change it into scalar matrix and identity matrix.

```python
arr6=np.array([[(int(input(f"Enter element at ({i},{j}) : "))) for j in range(3)] for i in range(3)])
print("\nEntered 2-D Array : \n", arr6)
i,j=0,0
while (j<3):
  arr6[j][i]=1
  i+=1
  if (i==3):
    j+=1
    i=0
print("\nConverted into Scalar Matrix : \n", arr6)
i,j=0,0
while (j<3):
  if (i==j):
    arr6[j][i]=1
  else:
    arr6[j][i]=0
  i+=1
  if (i==3):
    j+=1
    i=0
print("\nConverted into Identity Matrix : \n", arr6)
```

##7. Explain Inferential statistics and Descriptive statistics.

###Ans: Descriptive statistics summarizes the main features of a dataset and Inferential statistics makes predictions or draws conclusions about a population based on the observed sample data.
<ol>
  <li><b>Descriptive Statistics Techniques:
        <ul>
            <li>Measures of central tendency (mean, median, mode): Provide a central or typical value of a dataset.</li>
            <li>Measures of variability (range, variance, standard deviation): Indicate how spread out the values in a dataset are.</li>
            <li>Frequency distributions: Display the number of occurrences of different values in a dataset.</li>
            <li>Graphical representations (histograms, box plots, etc.): Visualize the distribution of data.</li>
        </ul>
  </li>
  <li><b>Inferential Statistics Techniques:
        <ul>
            <li>Hypothesis testing: Evaluates a hypothesis about a population parameter using sample data.</li>
            <li>Confidence intervals: Provides a range of values within which a population parameter is likely to fall.</li>
            <li>Regression analysis: Examines the relationship between variables and predicts the value of one variable based on another.</li>
            <li>Analysis of variance (ANOVA): Compares means of different groups to determine if there are significant differences.</li>
